```python
# load cliend data
clientList = []

# balancing classes
#oversample = RandomOverSampler(sampling_strategy='auto') #minority
smoteOversampling = SMOTE(sampling_strategy='auto',random_state=32)
#undersample = RandomUnderSampler(sampling_strategy='majority')

for folder in datasets:
    print (" ", str(inputFolders)+"student_"+folder+"_transformed.csv") #_numeric
    # load client data
    dataset = pd.read_csv(inputFolders+"student_"+folder+"_transformed.csv")
    
    # revome text date
    
    if 'timestamp_text' in dataset.columns:
        del  dataset['timestamp_text']
    
    dataset['class'] = dataset['class'].replace(['awake'], '0')
    dataset['class'] = dataset['class'].replace(['asleep'], '1')
    
    # print(dataset)
    y_train = dataset['class'].copy()
    print("initial len: ", dataset.shape)

    # does not add datasets that dont have instances from both classes
    if y_train.sum() != 0 and (y_train.sum() != len(y_train)):
        
        # do balancing technique
        X_train2, y_train2 = smoteOversampling.fit_resample(dataset, y_train)
        #X_train_under, y_train_under = undersample.fit_resample(X_train_under, y_train_under)
        
        dataset['class'] = y_train
        X_train2['class'] = y_train2
        
        clientList.append(dataset)
        
        dataset['class'].astype(int) 
        dataset['class'].astype(str) 
        dataset['class'] = dataset['class'].replace('0','awake')
        dataset['class'] = dataset['class'].replace('1','asleep' )
    
        # summarize data shape
        print("final len: ", X_train2.shape)
        print("----")
        # summarize class distribution
        print("Before oversampling: ", Counter(y_train))
        print("After undersampling: ", Counter(y_train2))
        dataset.to_csv(inputFolders+"student_"+folder+"_transformed_smote.csv", sep=',', encoding='utf-8', index=False)
        print("----")
        print("----")
        print("----")
        
print("Total",(len(clientList)))
```
